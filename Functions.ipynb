```python
def collatzlength(x,d):
    if x in d:
        return d[x]
    if x % 2 == 0:
        d[x] = 1 + collatzlength(x//2,d)
    else:
        d[x] = 1 + collatzlength(3*x+1,d)
    return d[x]

def argmaxcollatz(x):
    d = {1:0}
    maxi, pos = 0, 1
    for i in range(1,x):
        l = collatzlength(i,d)
        if maxi < l:
            maxi = l
            pos = i
    return pos
```

```python
argmaxcollatz(1000000)
```

```python
def isValid(y):
    if y < 1:
        return False
    if y == 1:
        return True
    d2 = False
    if y % 2 == 0:
        d2 = isValid(y//2)
    return d2 or isValid(y-3)
```

```python
isValid(262), isValid(646), isValid(800), isValid(61), isValid(44)
```

```python
def isConsecFibb(a,b):
    if a==1 and b==2:
        return True
    if a < 1:
        return False
    return isConsecFibb(a-b,a)
```

```python
def isPalindrome(s,start,end):
    if start >= end:
        return True
    if s[start] != s[end]:
        return False
    return isPalindrome(s,start + 1,end - 1)
```

```python
def findlargest(l):
    if len(l) == 1:
        return l[0]
    return max(l[0],findlargest(l[1:]))
```
